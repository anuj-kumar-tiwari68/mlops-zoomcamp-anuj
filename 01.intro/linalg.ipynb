```python
import numpy as np
u = np.array([1,2,3,4])
```

```python
v= np.array([5,6,7,8])
```

```python
u.shape
```

```python
v.shape
```

```python
def vector_vector_multiplication(u,v):
    assert u.shape[0] == v.shape[0]

    n = u.shape[0]
    result = 0.0
    for i in range(n):
        result =  result + u[i] * v[i]

    return result
```

```python
vector_vector_multiplication(u,v)
```

```python
u.dot(v)
```

```python
def metrix_vector_multiplication(U,v):
    assert U.shape[1]==v.shape[0]
    num_rows = U.shape[0]
    result = np.zeros(num_rows)
    for i in range(num_rows):
        result[i] = vector_vector_multiplication(U[i],v)
    return result
```

```python
metrix_vector_multiplication(U,v)
```
